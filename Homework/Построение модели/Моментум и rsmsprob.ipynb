```python
import numpy as np
```

```python
N = 100
X = np.random.uniform(low=0, high=100, size=N)
Y = 2*X + 1 + np.random.normal(scale=5, size=N)
```

```python
print(X)
```

```python
EPOCHS = 100
LEARNING_RATE = 0.0001
```

```python
def cost_function(X, y, theta0, theta1,gamma,vt0,vt1):
    total_cost = 0
    for i in range(len(X)):
        total_cost += ((theta0 + gamma * vt0) + (theta1 - gamma * vt1)*X[i] - y[i]) ** 2
    return total_cost / (2 * len(X))
```

```python
def der_theta0(X, y, theta0, theta1, gamma, vt0, vt1):
    total_cost = 0
    for i in range(len(X)):
        total_cost += ((theta0 + gamma * vt0) + (theta1 - gamma * vt1) * X[i] - y[i])
    return total_cost / (len(X))
```

```python
def der_theta1(X, y, theta0, theta1, gamma, vt0, vt1):
    total_cost = 0
    for i in range(len(X)):
        total_cost += ((theta0 - gamma * vt0) + (theta1 - gamma * vt1) * X[i] - y[i]) * X[i]
    return total_cost / (len(X))
```

```python
# Nesterov momentum
gamma = 0.1
t = 0
vt0 = 0
vt1 = 0
theta0 = 1
theta1 = 1
for _ in range(EPOCHS):
    dt0 = der_theta0(X, Y, theta0, theta1,gamma, vt0, vt1)
    dt1 = der_theta1(X, Y, theta0, theta1, gamma, vt0, vt1)
    if t==0:
        #нет gamma так как нет предыдущего значения vt
        theta0 = theta0 - LEARNING_RATE * dt0
        theta1 -= LEARNING_RATE * dt1
        vt0 = theta0
        vt1 = theta1
    else:
        theta0 = theta0 - LEARNING_RATE * dt0
        theta1 -= LEARNING_RATE * dt1
        vt0 += gamma*vt0 + theta0
        vt1 += gamma*vt1 + theta1
    print("t0:", vt0, "t1:", vt1, "cost:", cost_function(X, Y, theta0, theta1,gamma,vt0,vt1))
```
